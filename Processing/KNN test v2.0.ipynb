```python
import numpy as np
from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd

#reading training data
df = pd.read_csv('/Users/akshaypala/Documents/SYDE 462/ConsolidatedExpNov.csv')
#reading testing data
df2 = pd.read_csv('/Users/akshaypala/Documents/SYDE 462/Expert16.csv')
```

```python
X = np.array(df.drop(['class'], 1))
#label array
y = np.array(df['class'])

#test data array
test_data = np.array(df2.drop(['class'], 1))
print test_data
```

```python
#creating training and testing samples
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
```

```python
#define the classifier
clf = neighbors.KNeighborsClassifier()
```

```python
#train the classifier
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

```python
#defining testing data from expert16.csv
example_measures = test_data
```

```python
#predicting if testing data from expert16.csv is expert or not (should be expert)
example_measures = np.array(test_data)
example_measures = example_measures.reshape(len(example_measures), -1)
prediction = clf.predict(example_measures)
print(prediction)
```

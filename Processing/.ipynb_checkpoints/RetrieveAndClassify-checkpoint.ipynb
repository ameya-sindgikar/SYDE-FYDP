```python
import pymongo
import numpy as np

#Connect to mongoDB and get data from collection
client = pymongo.MongoClient() 
db = client.TrainingDataDB 
expCollection = db['XYexp2']
novCollection = db['XYnov2']
expDocuments = expCollection.find({})
novDocuments = novCollection.find({})
numExpDocuments = expDocuments.count()
numNovDocuments = novDocuments.count()
print numExpDocuments, numNovDocuments

currentExpData = []
for document in expDocuments:
    dataArray = document['data']
    currentExpData.append(dataArray)  
#print currentExpData
#print '-----------------'
currentNovData = []
for document in novDocuments:
    dataArray = document['data']
    currentNovData.append(dataArray)  
#print currentNovData


#TODO
# Iterate over all samples for the maneuver
```

```python
ed = np.array(currentExpData)
nd = np.array(currentNovData)
```

```python
np.array(currentNovData)
```

```python
from sklearn.manifold import TSNE

data = np.concatenate((ed, nd), axis=0)
labels = [0]*len(ed) + [1]*len(nd)
np.array(labels)

X = data
X.shape

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
X_tsne.shape
```

```python
import matplotlib.pyplot as plt

plt.title('t-SNE on Raw Exp and Nov Data')
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)
plt.show()
```

```python
extracted_expert_data = np.concatenate((np.mean(ed, axis=0), np.std(ed, axis=0)), axis=0)
extracted_novice_data = np.concatenate((np.mean(nd, axis=0), np.std(nd, axis=0)), axis=0)
```

```python
np.array(extracted_expert_data)
```

```python
extracted_data = np.concatenate((extracted_expert_data, extracted_novice_data), axis=0)

ext_labels = [0]*len(extracted_expert_data) + [1]*len(extracted_novice_data)
np.array(ext_labels)
```

```python
from sklearn.manifold import TSNE

ext_TSNE = TSNE(n_components=2)
Z_tsne = ext_TSNE.fit_transform(extracted_data)
Z_tsne.shape
import matplotlib.pyplot as plt

plt.title('t-SNE on Extracted Exp and Nov Data')
plt.scatter(Z_tsne[:, 0], Z_tsne[:, 1], c=ext_labels)
plt.show()
```
